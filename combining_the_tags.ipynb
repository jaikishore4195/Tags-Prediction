```python
import pickle
import pandas as pd
```

```python
with open("test_predictions.pkl", "rb") as fp:
    pred = pickle.load(fp)
```

```python
with open("parrot.pkl", "rb") as fp:
    new = pickle.load(fp)
```

```python
for i in range(len(new)):
    pred[i] = pred[i] + new[i]
    pred[i] = list(set(pred[i]))
```

```python
for i in range(0,len(pred)):
    temp=''
    for j in range(0,len(pred[i])):
        temp=temp+'|'+pred[i][j]
    pred[i]=temp[1:]
```
